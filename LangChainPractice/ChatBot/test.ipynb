```python
from langchain_community.vectorstores import Chroma
from langchain_openai import OpenAIEmbeddings
import os
from dotenv import load_dotenv

load_dotenv()
embeddings=OpenAIEmbeddings(api_key=os.environ['OPENAI_API_KEY'])
```

```python
vec_db = Chroma(persist_directory="./chroma_db", embedding_function=embeddings)
retriever = vec_db.as_retriever(search_type="similarity", search_kwargs={"k": 5})
request = 'What are the licenses for tableau?'
context = "\n".join(doc.page_content for doc in retriever.invoke(request))

print(context)
```
